# Base de Datos

```python
data = pd.read_csv('taylor_swift_spotify.csv') 
data
```

Se visualizan las columnas para verificar la informacion que se utilizÃ³, de esta manera se decidiÃ³ eliminar la columna "Unnamed: 0"

```python
data.columns
```

```python
tay_data = data.drop(columns=['Unnamed: 0'])
tay_data
```

La columna "duration_ms", como su nombre lo indica, muestra la duraciÃ³n de las canciones en milisegundos por lo que se claculan los datos de la columna en segundo y minutos.

```python
tay_data['duration_ms']
```

```python
tay_data['duration_ms'].dtypes
```

```python
tay_data['duration_ms']/1000
```

```python
tay_data['duration_s'] = tay_data['duration_ms']/1000
tay_data['duration_m'] = tay_data['duration_s']/60
tay_data
```

```python
tay_data.dtypes
```

Se realiza el cambio del tipo de dato para las siguentes columnas:

```python
tay_data['release_date'] = pd.to_datetime(tay_data['release_date'])
tay_data['name']=tay_data['name'].astype('string')
tay_data['album']=tay_data['album'].astype('string')
```

```python
tay_data.dtypes
```

Para esta ultima parte de la limpieza de datos, se decidio cambiar el nombre de las columnas y eliminar las columnas que no se van a necesitar para el analisis

```python
tay_data_rename_object = {
    'name' : 'Nombre',
    'album': 'Album', 
    'release_date': 'Fecha de lanzamineto',  
    'track_number': 'Numero de pista', 
    'acousticness':'Acustica', 
    'danceability':'Bailabilida', 
    'energy':'EnergÃ­a', 
    'instrumentalness':'Instrumentalidad',     
    'liveness': 'En Vivo', 
    'loudness':'Sonoridad', 
    'speechiness': 'Letra', 
    'tempo': 'Tempo', 
    'valence':'Valencia', 
    'popularity': 'Popularidad',
    'duration_m': 'Duracion (min)'}
tay_data.rename(columns=tay_data_rename_object, inplace=True, errors='raise')
```

```python
tay_data.columns
```

```python
new_tay_data = tay_data.drop(columns=['id','uri','duration_ms','duration_s']) #Se eliminan las columnas
new_tay_data
```

```python
new_tay_data.dtypes
```

Se guarda el dataset

```python
new_tay_data.to_csv('ts_data.csv', index=False)
```
